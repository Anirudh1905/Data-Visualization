```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Topic_Survey_Assignment.csv')
df.set_index(['Unnamed: 0'],inplace=True)
del(df.index.name)
df
```

```python
df.sort_values(['Very interested'],ascending=False,inplace=True)
df.sum(axis=1)
```

```python
for i in range(6):
    for j in range(3):
        df.iloc[i,j]=round(((df.iloc[i,j]/2233)*100),2)
df
```

```python
ax=df.plot(kind='bar',color=['#5cb85c','#5bc0de','#d9534f'],
           figsize=(20,8),width=0.8,fontsize=14)
ax.set_title('Percentage of Respondents Interest in Data Science Areas',fontsize=16)

ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.axes.get_yaxis().set_visible(False)
ax.legend(fontsize=14)
ax.tick_params(labelsize=14)

for p in ax.patches:
    ax.annotate(str(p.get_height()),(p.get_x()+p.get_width()/2,p.get_height()+2),ha='center',va='center',fontsize=14)

plt.show()
```
