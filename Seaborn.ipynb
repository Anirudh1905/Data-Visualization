```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Scatter Plot

```python
a=sns.load_dataset('flights')
sns.relplot(x='passengers',y='month',data=a)
```

```python
a=sns.load_dataset('flights')
sns.relplot(x='passengers',y='month',hue='year',data=a)
```

# Line Plot

```python
b=sns.load_dataset('tips')
sns.relplot(x='time',y='total_bill',hue='tip',data=b,kind='line')
```

# Catplot

```python
sns.catplot(x='day',y='total_bill',data=b)
```

# ViolinPlot 

```python
sns.catplot(x='day',y='total_bill',data=b,kind='violin')
```

# BoxenPlot

```python
sns.catplot(x='day',y='total_bill',data=b,kind='boxen')
```

## Bivariate

```python
x=pd.DataFrame({'Day':[1,2,3,4,5,6,7],'Grocery':[30,80,45,23,51,46,76],'Clothes':[13,40,34,23,54,67,98]})
y=pd.DataFrame({'Day':[8,9,10,11,12,13,14],'Grocery':[30,80,45,23,51,46,76],'Clothes':[13,40,34,23,54,67,98]})
mean,cov=[0,1],[(1,.5),(.5,1)]
data=np.random.multivariate_normal(mean,cov,200)
sns.jointplot(x=x,y=y,color='r')
```

# Jointplot-hex

```python
iris=sns.load_dataset("iris")
sns.jointplot("sepal_width","petal_length",data=iris,kind="hex",color='g')
```

## Jointplot-kde

```python
sns.jointplot("sepal_width","petal_length",data=iris,kind='kde',color='b')
```

## FacetGrid

```python
c=sns.FacetGrid(iris,col="species")
c.map(plt.hist,"sepal_length",color="orange")
```
